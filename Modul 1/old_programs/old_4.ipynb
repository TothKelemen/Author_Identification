```python
import pandas as pd
import os
import xml.etree.ElementTree as ET
import numpy as np
#import sys
#print(sys.version)
```

```python
def getAuthorName(poems_path):
    
    try:
    
        df0 = pd.read_xml(poems_path, xpath="//ns:titleStmt/ns:author/ns:persName", namespaces={"ns":"http://www.tei-c.org/ns/1.0"})

        Author=df0.iloc[0, 0] + ' ' + df0.iloc[0, 1]

        AuthorDict={
            'Author':Author
        }

        return AuthorDict
    
    except Exception as e:
        AuthorDict={
            'Author':np.nan
        }
        return AuthorDict
```

```python
def getDiv(poems_path):
    
    try:

        divKeyList=['Title','numWordsTitle','numStanza','numLine','numWord','numSyll','rateShortSyll']

        df1 = pd.read_xml(poems_path, xpath="//div")

        special_characters=['!',':',';','"','-','.','?',',','(',')','[',']','»','«','ʼ']

        Title=df1.iloc[0, 9]
        numWordsTitle=len(Title.replace(f"[{''.join(special_characters)}]",'').split())
        numStanza=df1.iloc[0, 1]
        numLine=df1.iloc[0, 2]
        numWord=df1.iloc[0, 3]
        numSyll=df1.iloc[0, 4]
        rateShortSyll=round(df1.iloc[0, 5] / numSyll,3)

        divValueList=[Title,numWordsTitle,numStanza,numLine,numWord,numSyll,rateShortSyll]

        divDict=dict(zip(divKeyList,divValueList))

        return divDict

    except Exception as e:
        divDict=dict.fromkeys(divKeyList,np.nan)
        return divDict
```

```python
def getStats(df,columnIndex):
    
    meanVar=round(df.iloc[:,columnIndex].mean(),3)
    maxVar=round(df.iloc[:,columnIndex].max(),3)
    minVar=round(df.iloc[:,columnIndex].min(),3)
    medVar=round(df.iloc[:,columnIndex].median(),3)
    q1Var=round(df.iloc[:,columnIndex].quantile(0.25),3)
    q3Var=round(df.iloc[:,columnIndex].quantile(0.75),3)
    vVar=round(df.iloc[:,columnIndex].std()/meanVar,3)
    
    statList=[meanVar,maxVar,minVar,medVar,q1Var,q3Var,vVar]
    
    return statList
```

```python
def getLg(poems_path):
    
    lgKeyList=['meanNL','maxNL','minNL','medNL','q1NL','q3NL','vNL',
                'meanNW','maxNW','minNW','medNW','q1NW','q3NW','vNW',
                'meanNS','maxNS','minNS','medNS','q1NS','q3NS','vNS',
                'meanNshortS','maxNshortS','minNshortS','medNshortS','q1NshortS','q3NshortS','vNshortS',
                'meanNlongS','maxNlongS','minNlongS','medNlongS','q1NlongS','q3NlongS','vNlongS',
                'abcb','aba','abcdb','abca','abb','aaaa'
               ]
    
    try:

        df2 = pd.read_xml(poems_path, xpath="//lg")    

        meanNL,maxNL,minNL,medNL,q1NL,q3NL,vNL=getStats(df2,1)          

        meanNW,maxNW,minNW,medNW,q1NW,q3NW,vNW=getStats(df2,2)

        meanNS,maxNS,minNS,medNS,q1NS,q3NS,vNS=getStats(df2,3)

        meanNshortS,maxNshortS,minNshortS,medNshortS,q1NshortS,q3NshortS,vNshortS=getStats(df2,4)

        meanNlongS,maxNlongS,minNlongS,medNlongS,q1NlongS,q3NlongS,vNlongS=getStats(df2,5)

        abcb=(df2.iloc[:, 6]=='abcb').sum()
        aba=(df2.iloc[:, 6]=='aba').sum()
        abcdb=(df2.iloc[:, 6]=='abcdb').sum()
        abca=(df2.iloc[:, 6]=='abca').sum()
        abb=(df2.iloc[:, 6]=='abb').sum()
        aaaa=(df2.iloc[:, 6]=='aaaa').sum()

        lgValueList=[meanNL,maxNL,minNL,medNL,q1NL,q3NL,vNL,
                     meanNW,maxNW,minNW,medNW,q1NW,q3NW,vNW,
                     meanNS,maxNS,minNS,medNS,q1NS,q3NS,vNS,
                     meanNshortS,maxNshortS,minNshortS,medNshortS,q1NshortS,q3NshortS,vNshortS,
                     meanNlongS,maxNlongS,minNlongS,medNlongS,q1NlongS,q3NlongS,vNlongS,
                     abcb,aba,abcdb,abca,abb,aaaa
                    ]

        lgDict=dict(zip(lgKeyList,lgValueList))

        return lgDict

    except Exception as e:
        lgDict=dict.fromkeys(lgKeyList,np.nan)
        return lgDict
```

```python
def getL(poems_path):
    
    lKeyList=['LmeanNW','LmaxNW','LminNW','LmedNW','Lq1NW','Lq3NW','LvNW',
               'LmeanNS','LmaxNS','LminNS','LmedNS','Lq1NS','Lq3NS','LvNS',
               'LmeanNshortS','LmaxNshortS','LminNshortS','LmedNshortS','Lq1NshortS','Lq3NshortS','LvNshortS',
               'LmeanNlongS','LmaxNlongS','LminNlongS','LmedNlongS','Lq1NlongS','Lq3NlongS','LvNlongS'
              ]
    
    try:
    
        df3 = pd.read_xml(poems_path, xpath="//l")

        LmeanNW,LmaxNW,LminNW,LmedNW,Lq1NW,Lq3NW,LvNW=getStats(df3,1)

        LmeanNS,LmaxNS,LminNS,LmedNS,Lq1NS,Lq3NS,LvNS=getStats(df3,2)

        LmeanNshortS,LmaxNshortS,LminNshortS,LmedNshortS,Lq1NshortS,Lq3NshortS,LvNshortS=getStats(df3,3)

        LmeanNlongS,LmaxNlongS,LminNlongS,LmedNlongS,Lq1NlongS,Lq3NlongS,LvNlongS=getStats(df3,4)

        lValueList=[LmeanNW,LmaxNW,LminNW,LmedNW,Lq1NW,Lq3NW,LvNW,
               LmeanNS,LmaxNS,LminNS,LmedNS,Lq1NS,Lq3NS,LvNS,
               LmeanNshortS,LmaxNshortS,LminNshortS,LmedNshortS,Lq1NshortS,Lq3NshortS,LvNshortS,
               LmeanNlongS,LmaxNlongS,LminNlongS,LmedNlongS,Lq1NlongS,Lq3NlongS,LvNlongS
              ]

        lDict=dict(zip(lKeyList,lValueList))

        return lDict

    except Exception as e:
        lDict=dict.fromkeys(lKeyList,np.nan)
        return lDict
```

```python
def getW(poems_path):
    
    wKeyList=['PROPN','NOUN','ADJ','NUM','PRON','VERB','ADV','CONJ','SCONJ','DET','ADP','INTJ','PART',
               'ratePROPN_NOUN','rateADJ','rateNOUNs','rateVERB','rateADV',
               'ratePhonTlow','ratePhonThigh','ratePhonTmixed',
               'meanLenWord','maxLenWord','minLenWord','medLenWord','q1LenWord','q3LenWord','vLenWord',
               'sumLenWord',
               'rateConsonant','rateShortFFV','rateLongFFV','rateShortBFV','rateLongBFV'
              ]
    
    try:
    
        df1 = pd.read_xml(poems_path, xpath="//div")

        numWord=df1.iloc[0, 3]

        df4 = pd.read_xml(poems_path, xpath="//w")

        PROPN=(df4.iloc[:, 2]=='PROPN').sum()
        NOUN=(df4.iloc[:, 2]=='NOUN').sum()
        ADJ=(df4.iloc[:, 2]=='ADJ').sum()
        NUM=(df4.iloc[:, 2]=='NUM').sum()
        PRON=(df4.iloc[:, 2]=='PRON').sum()
        VERB=(df4.iloc[:, 2]=='VERB').sum()
        ADV=(df4.iloc[:, 2]=='ADV').sum()
        CONJ=(df4.iloc[:, 2]=='CONJ').sum()
        SCONJ=(df4.iloc[:, 2]=='SCONJ').sum()
        DET=(df4.iloc[:, 2]=='DET').sum()
        ADP=(df4.iloc[:, 2]=='ADP').sum()
        INTJ=(df4.iloc[:, 2]=='INTJ').sum()
        PART=(df4.iloc[:, 2]=='PART').sum()

        ratePROPN_NOUN=round((PROPN+NOUN)/numWord,3)
        rateADJ=round(ADJ/numWord,3)
        rateNOUNs=round((PROPN+NOUN+ADJ+NUM+PRON)/numWord,3)
        rateVERB=round(VERB/numWord,3)
        rateADV=round(ADV/numWord,3)

        ratePhonTlow=round((df4.iloc[:, 5]=='low').sum()/numWord,3)
        ratePhonThigh=round((df4.iloc[:, 5]=='high').sum()/numWord,3)
        ratePhonTmixed=round((df4.iloc[:, 5]=='mixed').sum()/numWord,3)

        df4['lenWord']=df4.iloc[:, 6].str.len()

        meanLenWord,maxLenWord,minLenWord,medLenWord,q1LenWord,q3LenWord,vLenWord=getStats(df4,8)

        sumLenWord=df4.iloc[:, 8].sum()

        rateConsonant=round(df4.iloc[:, 6].str.count('c').sum()/sumLenWord,3)
        rateShortFFV=round(df4.iloc[:, 6].str.count('f').sum()/sumLenWord,3)
        rateLongFFV=round(df4.iloc[:, 6].str.count('F').sum()/sumLenWord,3)
        rateShortBFV=round(df4.iloc[:, 6].str.count('b').sum()/sumLenWord,3)
        rateLongBFV=round(df4.iloc[:, 6].str.count('B').sum()/sumLenWord,3)

        wValueList=[PROPN,NOUN,ADJ,NUM,PRON,VERB,ADV,CONJ,SCONJ,DET,ADP,INTJ,PART,
               ratePROPN_NOUN,rateADJ,rateNOUNs,rateVERB,rateADV,
               ratePhonTlow,ratePhonThigh,ratePhonTmixed,
               meanLenWord,maxLenWord,minLenWord,medLenWord,q1LenWord,q3LenWord,vLenWord,
               sumLenWord,
               rateConsonant,rateShortFFV,rateLongFFV,rateShortBFV,rateLongBFV
              ]    

        wDict=dict(zip(wKeyList,wValueList))

        return wDict
    
    except Exception as e:
        wDict=dict.fromkeys(wKeyList,np.nan)
        return wDict
```

```python
def getPC(poems_path):
    
    try:
        
        df5 = pd.read_xml(poems_path, xpath="//pc")

        PUNCT=(df5.iloc[:, 1]=='PUNCT').sum()

        PUNCT_Dict={
            'PUNCT':PUNCT
        }

        return PUNCT_Dict
    
    except Exception as e:
        PUNCTdict={
            'PUNCT':np.nan
        }
        return PUNCTdict
```

```python
def getRhymePair(poems_path):
    
    try:

        df6 = pd.read_xml(poems_path, xpath="//rhymePair")

        numRhymePairs=df6.iloc[:, 2].count()

        rhymePairDict={
            'numRhymePairs':numRhymePairs
        }

        return rhymePairDict
    
    except Exception as e:
        rhymePairDict={
            'numRhymePairs':np.nan
        }
        return rhymePairDict
```

```python
def getAlliteration(poems_path):
    
    allKeyList=['numAll','rateClearAll','meanAll','maxAll','minAll','medAll','q1All','q3All','vAll']
    
    try:
        
        df7 = pd.read_xml(poems_path, xpath="//alliteration")

        numAll=df7.iloc[:, 1].count()

        df7['lenAll']=df7.iloc[:, 1].apply(len)

        rateClearAll=round((df7.iloc[:, 1].str.count('a')==df7.iloc[:, 6]).sum()/numAll,3)

        meanAll,maxAll,minAll,medAll,q1All,q3All,vAll=getStats(df7,6)

        allValueList=[numAll,rateClearAll,meanAll,maxAll,minAll,medAll,q1All,q3All,vAll]
        
        AllDict=dict(zip(allKeyList,allValueList))

        return AllDict
    
    except Exception as e:
        AllDict=dict.fromkeys(allKeyList,np.nan)
        return AllDict
```

```python
# Kiindulási mappa, amiben a további almappák vannak
poemsPath="D:\Iskolai\Tananyagok\Szakdolgozat\Bsc\Versek"

rawIndex=0

df=pd.DataFrame()

# os.walk használata az összes almappa bejárásához
for rootFolder, Folder, Files in os.walk(poemsPath):
    for File in Files:
        if File.endswith('.xml'):  # Csak az XML kiterjesztésű fájlokat vesszük figyelembe
            completePath = os.path.join(rootFolder, File)

            # XML fájl beolvasása
            try:
                tree = ET.parse(completePath)
                root = tree.getroot()

                # Itt dolgozhatsz az XML tartalommal
                # Például: print(root.tag)

                AuthorDict=getAuthorName(completePath)

                divDict=getDiv(completePath)

                lgDict=getLg(completePath)

                lDict=getL(completePath)

                wDict=getW(completePath)

                pcDict=getPC(completePath)

                rhymePairDict=getRhymePair(completePath)

                allDict=getAlliteration(completePath)

                mainDict={**AuthorDict,**divDict,**lgDict,**lDict,**wDict,**pcDict,**rhymePairDict,**allDict}

                if rawIndex==0:                        

                    column_names=list(mainDict.keys())            

                    df = pd.DataFrame(columns=column_names)                   

                for item in mainDict:    

                    df.loc[rawIndex, item] = mainDict[item]

                rawIndex+=1

            except ET.ParseError:
                print(f'Hiba a(z) {File} XML fájl olvasásakor.')

#print(mainDictionary)
```
